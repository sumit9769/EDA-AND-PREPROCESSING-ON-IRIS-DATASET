```python
n1=np.array([1,2,3,4,5])
n2=np.array([10,20,30,40,50])
n3=np.array([100,200,300,400,500])
```

```python
# MIN-MAX SCALING
def minmax(arr):
    xnew=(arr-arr.min())/(arr.max()-arr.min())
    return xnew
```
